## Предобработка данных

В столбце **'last_price'** не обязательно иметь цыфру после запятой.

```python
data['last_price'] = data['last_price'].astype('int')
#проверка значений на странные значения и возможные ошибки 
```

В столбце **'first_day_exposition'** можно привести дату к типу **'datetime64'**. Так как в данных не указаны часы их можно не преобразовывать.

```python
data['first_day_exposition'].value_counts()
#проверка значений на странные значения и возможные ошибки 
```

```python
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format = '%Y-%m-%d')
#изменение типа данных стобца на значение 'datetime64'
```

```python
data['first_day_exposition'] #проверка типпа данных
```

В столбце **'ceiling_height'** пропущено 9195 значений, это 39% от всего датафрейма. В таких случаях лучше заполнить пропуски, что бы не потерять важные значения. Данный столбец отражает высота потолков в м. Эти пропуски можно заменить значением медианы. Так же в данном столбце присутствуют подозрительно высокие значения.

```python
data['ceiling_height']=data['ceiling_height'].fillna(data['ceiling_height'].median()) 
#замена порпусщиных значений на медиану
```

```python
data['ceiling_height'].value_counts() #проверка значений на странные значения и возможные ошибки 
```

```python
data['ceiling_height'].describe() #проверка значений столбца
```

```python
data[data['ceiling_height'] > 2.7].sort_values('ceiling_height').tail(10) #просмотр потенциально анамальных значений
```

В столбце **'floors_total'** отсутствует менее 1% данных, от всех значений в датафрейме, по этому пропущеные данные можно удалить. Так же можно изменить тип данных, так как число после запятой здесь не обязательно.

```python
data['floors_total'].value_counts() #проверка значений на странные значения и возможные ошибки 
```

```python
data = data.dropna(subset =['floors_total']).reset_index(drop=True) 
#удаление пропущеных хначений в стобце 'floors_total' и обновление индексов
```

```python
data['floors_total'] = data['floors_total'].astype('int') #изменение типа данных
```

```python
data['floors_total'].isna().sum() #проверка изменений
```

В столбце **'is_apartment'** отсутствует слишком много данных(88%). Скорее всего если не указано, что квартира является аппартаментами, то она ими и не является. Тогда логично заменить пропущенные значения на False.

```python
data['is_apartment']=data['is_apartment'].fillna(value=False) 
#замена порпусщиных значений на медиану
```

```python
data['is_apartment'].value_counts() #проверка значений на странные значения и возможные ошибки 
```

Данные в столбце **'kitchen_area'** отражают площадь кухни в квадратных метрах. В этом столбце отсутствует 2278 значений (10%).  Его можно вычислить через коэффициент.

```python
kitchen_coef = data['kitchen_area'].median() / data['total_area'].median() #нахождение коэффициента
```

```python
data['kitchen_area'].fillna(kitchen_coef * data['total_area'], inplace=True) #замена порпусщиных значений
```

```python
data['kitchen_area'].value_counts() #проверка значений на странные значения и возможные ошибки 
```

```python
data['kitchen_area'].isna().sum()#проверка изменений
```

В столбце **'living_area'** пропущено 1903 значений, это 8% от датафрейма. Его можно вычислить через коэффициент.

```python
living_coef = data['living_area'].median() / data['total_area'].median() #нахождение коэффициента
```

```python
data['living_area'].fillna(living_coef * data['total_area'], inplace=True) #замена порпусщиных значений
```

```python
data['living_area'].value_counts() #проверка значений на странные значения и возможные ошибки 
```

## Расчёты и добавление результатов в таблицу

### Добавление цены квадратного метра;


```python
data['price_area'] = data['last_price'] / data['living_area'] #вычисление цены квадратного метра
```

### Добавление дня недели, месяца и года публикации объявления;

```python
data['day_of_the_week'] = data['first_day_exposition'].dt.weekday #добавление столбца с днями недели
```

```python
data['month'] =  data['first_day_exposition'].dt.month #добавление столбца с месяцами
```

```python
data['year'] = data['first_day_exposition'].dt.year #добавление столбца с годами
```

### Добавление этажа квартиры; варианты — первый, последний, другой;

```python
def floor_category(rew): #создание функции котигоризирующей доход
    if rew['floor'] == 1:
        return 'первый'
    elif rew['floor'] == rew['floors_total']:
        return 'последний'
    else:
        return 'другой'
```

```python
data['floor_category'] = data.apply(floor_category, axis =1)  #запись получившихся котигорий в новый столбец
```

```python
data['floor_category'].value_counts() #проверка получившихся значенний
```

```python
data['ratio_kitchen_area'] = data['kitchen_area']/data['total_area'] #Нахождение соотношения площади кухни и общей площади
```

### Удаление редких и выбивающихся значений

В ходе анализа были выявлены столбцы с выбивающимеся значениями: 'total_area', 'kitchen_area', 'living_area', 'last_price', 'rooms', 'ceiling_height', 'days_exposition'. 

Что бы очистить датафрейм можно удалить значения по правилу  1.5IQR.

```python
def quartile_range(df, column):   # Функция для подсчёта границ с учетом 1.5 размаха
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    dfq = df.loc[(df[column] < q3 + 1.5*iqr) & (df[column] > q1 - 1.5*iqr), column]
    return dfq
```

```python
list_features = ['last_price', 'total_area', 'kitchen_area', 'living_area', 'days_exposition', 'ceiling_height', 'rooms']
for col in list_features:
    data[col] = quartile_range(data, col)
# Столбцы с выбросами
```

```python
data.describe() #изучение новых показателей
```

Так мы избавились от сомнительных данных.
